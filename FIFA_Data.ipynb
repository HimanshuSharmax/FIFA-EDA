```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
fifa = pd.read_csv('players_20.csv')
fifa.head()
```

```python
# Printing all columns names

for col in fifa.columns:
    print(col)
```

```python
fifa['nationality'].value_counts()
```

```python
fifa['nationality'].value_counts()[0:5]
```

```python
fifa['nationality'].value_counts()[0:5].keys()
```

```python
# bar plot | first value -> catalog, second value -> numerical

plt.figure(figsize=(5,5))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()), list(fifa['nationality'].value_counts()[0:5]))
plt.show()
```

```python
# Finding out player salary

player_salary = fifa[['short_name','wage_eur']]
player_salary.head()
```

```python
# player salary in desc order

player_salary = player_salary.sort_values(by=['wage_eur'], ascending = False)
player_salary.head()
```

```python
# Bar plot acc to player salary

plt.figure(figsize=(8,3))
plt.bar(list(player_salary['short_name'])[0:5], list(player_salary['wage_eur'])[0:5],
        color=['skyblue', 'lightcoral', 'lightgreen', 'gold', 'plum'] )
plt.show()
```
